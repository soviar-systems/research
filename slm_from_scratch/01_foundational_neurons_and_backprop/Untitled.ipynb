Write a Python function (using numpy) that takes an input vector x of shape (2,) and weights w of shape (2,) and a scalar bias b. It should return the scalar output of the linear sum.

```python
x = np.array([0, 2])
w = np.array([1, 1])
b = 3

def line_sum(x, w, b):
    return np.float64(np.dot(x, w) + b)
    
print(x.shape)
result = line_sum(x, w, b)
print(result)
print(result.shape)
```
